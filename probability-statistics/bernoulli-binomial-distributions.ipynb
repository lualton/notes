# Bernoulli

1. A single bernoulli trial is used to determine _success_ or _failure_. This is a coin flip, whether a team scores or not, if it will rain or not.


2. Parameters: 
    - P: Probability of success. A normal coin has p=0.5 where you flip a heads 50% of the time. p can also be considered the shape parameter because it determines the shape of the distribution
    
    
3. PMF
\begin{align}
P(p) = \left\{ \begin{array}{ccc}
1-p & & heads = 0 \\[0.5em]
p & & heads = 1 \\[0.5em]
\end{array}
\right.
\end{align}

4. When is it used?
    - Bernoulli Random Variables are like a base line random variable that are extended to many other distributions


```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-white')
import numpy as np
from scipy.stats import *
import seaborn as sns
```

```python
data_bern = bernoulli.rvs(size=1000,p=0.6)
ax = sns.distplot(data_bern,
                  kde=False,
                  color='crimson',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Bernouli', ylabel='Frequency');
```

```python
f, ax = plt.subplots(1, 3, sharey=True, figsize=(16, 8))
p = [0.5, 0.75, 0.95]

for idx, i in enumerate(p):
    data_bern = bernoulli.rvs(size=10000, p=i)
    ax1 = sns.distplot(data_bern,
                  kde=False,
                  color='crimson',
                  hist_kws={"linewidth": 25,'alpha':1},
                  ax=ax[idx])
    ax1.set_title("Bernoulli with P = " + str(i), fontsize = 15)
    ax1.set_xlabel("0 = Tails; 1 = Heads", fontsize = 12)
    ax1.set_ylabel("Frequency out of 10,000 trials", fontsize = 12)
```

#### Changing the P parameter changes how likely it is that we get a heads (or it rains, or whatever the condition of success is).

## Binomial

1. A binomial are just repeated bernoulli trials that calculate how often a success occurs.


2. Parameters: 
    - P: Probability of success. A normal coin has p=0.5 where you flip a heads 50% of the time. p can also be considered the shape parameter because it determines the shape of the distribution
    - n: The number of trials to run.
    
    - P = 0.5 and n = 10 means this binomial distribution takes the form of 10 coin flips. We would count the number of successes that occur across all trials. K is if you look for a specific number of successes
    
    
3. PMF
\begin{align}
P(k;n,p) = \begin{pmatrix}
n \\
k
\end{pmatrix}
p^k (1-p)^{n-k}.
\end{align}

4. When is it used?
    - Bernoulli Random Variables are like a base line random variable that are extended to many other distributions


```python
data_binom = binom.rvs(n=20,p=0.8,loc=0,size=1000)
ax = sns.distplot(data_binom,
                  kde=True,
                  color='blue',
                  hist_kws={"linewidth": 25,'alpha':1})
ax.set(xlabel='Binomial', ylabel='Frequency');
```

```python
f, ax = plt.subplots(1, 3, sharey=True, figsize=(16, 8))
p = [0.5, 0.75, 0.95]
n = [100, 1000, 10000]

for idx, i in enumerate(p):
    data_binom = binom.rvs(n=n[idx], p=i,loc=0,size=1000)
    ax1 = sns.distplot(data_binom,
                  kde=False,
                  color='crimson',
                  hist_kws={"linewidth": 25,'alpha':1},
                  ax=ax[idx])
    ax1.set_title("Bernoulli with P = " + str(i), fontsize = 15)
    ax1.set_xlabel("0 = Tails; 1 = Heads", fontsize = 12)
    ax1.set_ylabel("Frequency out of 10,000 trials", fontsize = 12)
```
