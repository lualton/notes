## Start

```python
import matplotlib.pyplot as plt
import seaborn as sns

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
tips = sns.load_dataset('Tips')
cars = sns.load_dataset('mpg')
```

```python
tips.head()
```

```python
cars.head()
```

### Scatterplot

```python
sns.scatterplot(x='total_bill', y='tip', data = tips,
                hue='sex',
                hue_order=['Male', 'Female'],
                palette='Set2')
plt.show()
```

### Countplot

```python
sns.countplot(x='sex', data = tips, hue='smoker')
plt.show();
```

### Relational Plots

- Useful for subplots
- Additional scatter/line plot

```python
# change col to row for row based plots
sns.relplot(x="total_bill", y="tip", data=tips, kind="scatter", 
            col= "smoker",
           col_order=['Yes', 'No'],
           row = 'sex',
           row_order = ['Female', 'Male'])
plt.show();
```

```python
sns.relplot(x="model_year", y="horsepower", 
            data=cars, kind="line", 
            ci=None, style="origin", 
            hue="origin", markers=True)

# Show plot
plt.show()
```

## Customize colors, point style, alpha

```python
sns.relplot(x='total_bill', y = 'tip', data = tips,
           kind='scatter', 
           size = 'size',
           hue = 'size',
           style = 'smoker',
           alpha = 0.8)
plt.show();
```

## Categorical Plots (Count and Bar plots)

- Comparisons between different groups

### Barplot

```python
sns.catplot(x="origin", y="mpg",
            data=cars,
            kind="bar",
            order = ['japan', 'europe', 'usa'],
            ci=None)

# Show plot
plt.show();
```

### Boxplot

```python
# Box plot
# sym='' omits outliers

sns.catplot(x="origin", y="mpg",
            data=cars,
            kind="box",
            order = ['japan', 'europe', 'usa'],
            ci=None)

# Show plot
plt.show();
```

### Pointplot

```python
# Means
# Remove the lines joining the points
sns.catplot(x="sex", y="tip",
            data=tips,
            # The kind of plot
            kind="point",
            # Horizonal 'cap' line at the ends
            capsize=0.2,
            # Have connecting lines to compare
            join=False)
            
# Show plot
plt.show()
```

```python
# Median

# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="smoker", y="total_bill",
			data=tips,
            kind="point",
            hue="sex",
            ci=None,
            estimator=median)

# Show plot
plt.show()
```

## Titles/Labels

Key information:
- relplot and catplot create FacetGrid object types
- scatterplot, countplot, etc. create AxesSubplot types


```python
g = sns.relplot(x='total_bill', y='tip', data = tips,
                hue='sex', kind='scatter', row='time')
g.fig.suptitle('Male vs Female Tipping', y = 1.03)
g.set_titles('Tipping during {row_name}')
g.set(xlabel = 'Total Bill',
      ylabel = 'Tips')
plt.show();
```

```python
g = sns.scatterplot(x='total_bill', y='tip', data = tips,
                hue='sex')
g.set_title("Title Here")
g.set(xlabel = 'Total Bill',
      ylabel = 'Tips')

plt.show();
```
