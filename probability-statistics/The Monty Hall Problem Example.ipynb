### The Monty Hall Problem


This is a famous probability puzzle written in the American Statistican in 1975. https://en.wikipedia.org/wiki/Monty_Hall_problem

The general structure is that you are on a game show with three doors. One door has a car and the other two have goats. 

You begin by choosing a door. THe host (Monty) will then reveal a goat from the two doors you didn't pick. Monty will then ask "Would you like to switch your choice?" To maximize your chance of winning, you should switch doors.

### Define a function for the monty hall game

```python
# Function to simulate one monty hall run
def monty_hall():
    '''
    returns whether you win by switching or not. Depending on what you put in.
    
    '''
    cardoor = random.randint(0,2)
    pickdoor = random.randint(0,2)
    
    if cardoor == pickdoor:
        return('stay wins')
    
    if cardoor != pickdoor:
        return('switch wins')
```

### Create a list to contain results and simulate the game

```python
# Initialize a list of wins and however many times you want to run the simulation
wins= []
number_runs = range(5000)
```

```python
for i in number_runs:
    wins.append(monty_hall())
```

### Calculate win percentages

```python
print("Staying with initial door wins " + str(wins.count('stay wins')/len(wins)) + "% of the time")
print("Switching your door wins " + str(wins.count('switch wins')/len(wins)) + "% of the time")
```

```python
plt.hist(wins)
plt.ylabel('Number of Wins');
```
